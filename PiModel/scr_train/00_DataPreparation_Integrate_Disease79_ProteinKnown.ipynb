```python
import sys  
import torch
import random
import argparse
import numpy as np
import time
import torch.nn as nn
import pandas as pd
import os

# # Import the required files
# sys.path.insert(1, '../scr_train/')
# import utils, ConcatNN, train_gene_nn
# from utils import *
# from ConcatNN import *
# from train_gene_nn import *

# Apply the seed to reproduct the results
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)


# =================================================================================
# Default settings
parser = argparse.ArgumentParser()

# =================================================================================
# Protein signatures.
# =================================================================================
parser.add_argument('--pert_type', type=str, default=pert_type, help='Perturbation type of protein signatures, e.g., trt_sh.cgs or trt_oe') # knockdown signatures
# parser.add_argument('--pert_type', type=str, default='trt_oe', help='Perturbation type of protein signatures, e.g., trt_sh.cgs or trt_oe') # overexpression signatures
parser.add_argument('--cell_name', type=str, 
                    default=cell_name, 
                    help='Cell name of LINCS files, e.g., mcf7') # cell line of protein signatures
parser.add_argument('--target_select', type=str, default='Known', 
                    help='Targets used for training. If supervised learning, only known proteins are used for training.')
parser.add_argument('--target_feature_type', type=str, default= target_feature_type, 
                    help='target feature types, e.g., Epo2000_Lr0.0001_Hid512_256_128_Lat64_Bat64_Dro0.1_ActTanh, VarianceTop1 or Original')
parser.add_argument('--target_expression_file', type=str, default='../../../LINCS/latest_version/imputation/imputed_data/', 
                    help='Path of the training gene expression profile dataset for the NN')
parser.add_argument('--target_vae_type', type=str, default=target_vae_type, help='Protein VAE type, e.g., VAE, VQ_VAE or Original')

# Select Cell lines based on missing rates.
# --cell_name='AllCell'の時のみ使用
parser.add_argument(
    '--cell_missing_rate', type=float, 
    default=missing_rate,
    help='Threshold of missing rates of cell lines when selecting AllCellParallel as --cell_name'
)

# =================================================================================
# Disease signatures.
# =================================================================================
parser.add_argument('--disease_profile_type', type=str, default=disease_profile_type, 
                    help='Disease signature type, e.g., disease_signature.tabs or patient_signature.tabs') # disease signatures
parser.add_argument('--disease_select', type=str, default='79', 
                    help='Diseases used for training. If supervised learning, only known diseases are used for training.')
parser.add_argument('--disease_feature_type', type=str, default=disease_feature_type, 
                    help='disease feature types, e.g., Epo2000_Lr0.0001_Hid512_256_128_Lat64_Bat64_Dro0.2_ScaleStd, VarianceTop1 or Original')
parser.add_argument('--disease_expression_file', type=str, default='../../../CREEDS/processed_data/expression-based_', 
                    help='Path of the training gene expression profile dataset for the VAE')
parser.add_argument('--disease_vae_type', type=str, default=disease_vae_type, help='Disease VAE type, e.g., VAE, VQ_VAE or Original')

# =================================================================================
# Pairwise type
# =================================================================================
parser.add_argument('--pairwise_type', type=str, default='Concat', help='Feature types of protein–disease pairs, e.g., Concat or DisCosine_TarCosine_Kron')

# =================================================================================
# Hyper-parameters of GeneVAE
# =================================================================================
parser.add_argument("-f", required=False)
args = parser.parse_args()
```

```python
# ====== Make output directory ====== #

def MakeOutputDir(args):
    
    if args.disease_vae_type == 'Original' and args.target_vae_type == 'Original':
        disease_feature_path = f"DiseaseFeaeture_{args.disease_feature_type}"
        target_feature_path = f"TargetFeaeture_{args.target_feature_type}"
    else:
        disease_feature_path = f"DiseaseFeaeture_{args.disease_vae_type}_{args.disease_feature_type}"
        target_feature_path = f"TargetFeaeture_{args.target_vae_type}_{args.target_feature_type}"

    o_base_dir = f"../data/{args.pert_type}/{disease_feature_path}/" # Disease feature type
    o_base_dir = f"{o_base_dir}/{target_feature_path}/" # Target feature type
    o_base_dir = f"{o_base_dir}/CellLine_{args.cell_name}/PairD{args.disease_select}T{args.target_select}" # Cell line
    
    return o_base_dir
```

```python
def show_gene_vae_hyperparamaters(args):

    # Hyper-parameters
    params = {}
    print('\n\nConcatNN Hyperparameter Information:')
    print('='*50)
    params['PERTURBATION_TYPE'] = args.pert_type 
    params['CELL_NAME'] = args.cell_name
    params['TARGET_FEATURE_TYPE'] = args.target_feature_type
    
    params['DISEASE_PROFILE_TYPE'] = args.disease_profile_type
    params['DISEASE_FEATURE_TYPE'] = args.disease_feature_type
    
    params['DISEASE_VAE_TYPE'] = args.disease_vae_type
    params['PROTEIN_VAE_TYPE'] = args.target_vae_type
    
    params['DISEASE_UNLABELED_DATA'] = args.disease_select
    params['TARGET_UNLABELED_DATA'] = args.target_select

    for param in params:
        string = param + ' ' * (5 - len(param))
        print('{}:   {}'.format(string, params[param]))
    print('='*50)
    
    
# ========================================================= #
# Print GeneVAE hyperparameter information
show_gene_vae_hyperparamaters(args)
```

```python
# ============================================================================
# Select cell lines based on missing rates.
# ============================================================================
def select_clls_based_on_missing_rates(args, cell_missing_rate):

    # Missing rate data.
    i_f = "../../../target_repositioning2/multitask/data/LINCS/cell_kdoe_list.txt"
    miss_df = pd.read_csv( i_f, sep ='\t', encoding='shift-jis' )

    if args.pert_type == 'trt_sh.cgs':
        col_name = 'kd_missing'
    elif args.pert_type == 'trt_oe':
        col_name = 'oe_missing'

    # Select cell lines based on missing rate.
    miss_df = miss_df.dropna(subset=col_name) # Select overexpression signature's cells
    miss_df = miss_df[miss_df[col_name] >= cell_missing_rate ] # Select cell lines based on missing rate
    cell_lines = sorted(set(miss_df['cell'])) # Cell line list.

    return cell_lines


# ====== Cell Lines ====== #

all_cell_lines = select_clls_based_on_missing_rates(args, 0.0)
cell_lines = select_clls_based_on_missing_rates(args, 0.1)
print(len(cell_lines))

cell_lines
```

```python
# ============================================================================
# Protein data: 全細胞のデータをマージする
# ============================================================================

protein_data = pd.DataFrame()

if missing_rate == 0.0:
    for cell in all_cell_lines:

        # VAE features.
        if cell in cell_lines:
            tmp_feature_type = args.target_feature_type.replace('Missing0.0', 'Missing0.1')
            i_f = f'../../VAE/data/{args.target_vae_type}/{args.pert_type}/CellLine_{cell}/{tmp_feature_type}/latent_vectors.txt'
        else:
            tmp_feature_type = args.target_feature_type.replace('_Missing0.0', '')
            i_f = f'../../VAE/data/{args.target_vae_type}/{args.pert_type}/CellLine_{cell}/{tmp_feature_type}/latent_vectors.txt'
        tmp_data = pd.read_csv( i_f, sep = '\t', index_col='cmap_name')
        tmp_data.columns = [cell + '__' + str(s) for s in tmp_data.columns] # Add cell name to columns.

        tmp_data = tmp_data.loc[ gold_genes ] # known proteinsを選択、全ての細胞で並びを揃えるため
    
        protein_data = pd.concat([protein_data, tmp_data], axis=1) # merge
        
elif missing_rate == 0.1:
    for cell in cell_lines:
        
        # VAE features.
        i_f = f'../../VAE/data/{args.target_vae_type}/{args.pert_type}/CellLine_{cell}/{target_feature_type}/latent_vectors.txt'
        tmp_data = pd.read_csv( i_f, sep = '\t', index_col='cmap_name')
        tmp_data.columns = [cell + '__' + str(s) for s in tmp_data.columns] # Add cell name to columns.

        tmp_data = tmp_data.loc[ gold_genes ] # known proteinsを選択、全ての細胞で並びを揃えるため
    
        protein_data = pd.concat([protein_data, tmp_data], axis=1) # merge
        

# 分散に基づき遺伝子を選択
if 'Variance' in args.target_feature_type:
    protein_data = protein_data[tar_fea_list] # geneを選択

# 標準化
protein_data = (protein_data - protein_data.min()) / (protein_data.max() - protein_data.min()) # 各特徴量について標準化
protein_data = protein_data.dropna(how = 'any', axis = 1)

# Feature data.
protein_feature_df = pd.DataFrame(list(protein_data.columns), columns=['features']) # features used for training.
o_dir = f"{MakeOutputDir(args)}/feature_files/" # Cell line
os.makedirs( o_dir, exist_ok=True ) # Make dirs.
o_f = f"{o_dir}/protein_features.txt"
protein_feature_df.to_csv( o_f, sep = '\t', index = None)


print('Protein data: ', protein_data.shape) # (nums, dimension)
protein_data
```

```python
# ====== Disease data ======= #
# Read disease gene expression data, which contains disease ids and gene values

if args.disease_vae_type == 'Original':
    # Original features.
    all_disease_data = pd.read_csv(args.disease_expression_file + args.disease_profile_type, 
                             sep = '\t',
                             index_col='disease'
                            )
    
else:
    # VAE features.
    i_f = f'../../VAE/data/{args.disease_vae_type}/disease/{args.disease_profile_type}/{args.disease_feature_type}/latent_vectors.txt'
    all_disease_data = pd.read_csv(i_f, sep = '\t', index_col='disease')
    
l_disease_data = all_disease_data.loc[ l_dis_list ] # Select known diseases.

u_dis_list = list(set(all_disease_data.index) - set(l_dis_list)) # Unlabeled disease list
u_dis_list1 = [ s for s in u_dis_list if '_' not in s] # Select disease-specific list
u_dis_list2 = [s+'_01' for s in u_dis_list1] # Select patient 01 id list

u_disease_data1 = all_disease_data.loc[u_dis_list1] # Select unlabeled diseas signatures
u_disease_data2 = all_disease_data.loc[u_dis_list2] # Select unlabeled diseas signatures

# 遺伝子を選択
if 'Variance' in args.disease_feature_type:
    l_disease_data = l_disease_data[dis_fea_list] # Select genes.
    u_disease_data1 = u_disease_data1[dis_fea_list] # Select genes.
    u_disease_data2 = u_disease_data2[dis_fea_list] # Select genes.

# 標準化
l_disease_data = (l_disease_data - l_disease_data.min()) / (l_disease_data.max() - l_disease_data.min()) # 各特徴量について標準化
l_disease_data = l_disease_data.dropna(how = 'any', axis = 1)
u_disease_data1 = (u_disease_data1 - u_disease_data1.min()) / (u_disease_data1.max() - u_disease_data1.min()) # 各特徴量について標準化
u_disease_data1 = u_disease_data1.dropna(how = 'any', axis = 1)
u_disease_data2 = (u_disease_data2 - u_disease_data2.min()) / (u_disease_data2.max() - u_disease_data2.min()) # 各特徴量について標準化
u_disease_data2 = u_disease_data2.dropna(how = 'any', axis = 1)

# Feature data.
l_disease_feature_df = pd.DataFrame(list(l_disease_data.columns), columns=['features'])
u_disease_feature_df1 = pd.DataFrame(list(u_disease_data1.columns), columns=['features'])
u_disease_feature_df2 = pd.DataFrame(list(u_disease_data2.columns), columns=['features'])
o_dir = f"{MakeOutputDir(args)}/feature_files/" # Cell line
os.makedirs( o_dir, exist_ok=True ) # Make dirs.
o_f = f"{o_dir}/l_disease_features.txt"
l_disease_feature_df.to_csv( o_f, sep = '\t', index = False )
o_f = f"{o_dir}/u_disease_features1.txt"
u_disease_feature_df1.to_csv( o_f, sep = '\t', index = False )
o_f = f"{o_dir}/u_disease_features2.txt"
u_disease_feature_df2.to_csv( o_f, sep = '\t', index = False )

print('Labeled disease data: ', l_disease_data.shape) # (nums, dimension)
print('Unlabeled disease data1 : ', u_disease_data1.shape) # (nums, dimension)
print('Unlabeled disease data2 : ', u_disease_data2.shape) # (nums, dimension)
l_disease_data
```
