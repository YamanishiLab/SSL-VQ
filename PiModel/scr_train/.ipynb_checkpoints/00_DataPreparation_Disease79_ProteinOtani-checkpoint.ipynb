# Semi-supervised NN用にデータを準備

・5 fold cross validation用にデータを分割  
・ラベルなしサンプル：  
known proteins x orphan diseases  
orphan proteins（大谷さんのデータと大和さんのデータの差分） x known diseases

```python
# ===============================================================
# Perturbation type.
# ===============================================================
pert_type = 'trt_sh.cgs'
# pert_type = 'trt_oe'


# ===============================================================
# Cell line.
# ===============================================================
# cell_name = 'AllCell'
cell_name = 'AllCellParallel'
# cell_name = 'All'


# ===============================================================
# Protein VAE type.
# ===============================================================
# target_vae_type = 'VAE'
target_vae_type = 'VQ_VAE'
# target_vae_type = 'SQ_VAE'
# target_vae_type = 'GaussianSQVAE_gaussian_1'
# target_vae_type = 'Original' # originalのgene expression scoreを使用


# ===============================================================
# Disease VAE type.
# ===============================================================
# disease_vae_type = 'VAE'
disease_vae_type = 'VQ_VAE'
# target_vae_type = 'SQ_VAE'
# disease_vae_type = 'GaussianSQVAE_gaussian_1'
# disease_vae_type = 'Original' # originalのgene expression scoreを使用


# ===============================================================
# Unlabeled sample.
# ===============================================================
# Proteins.
# target_select = 'Known'
target_select = 'Otani'

# Diseases.
disease_select = '79'

# ===============================================================
# Missing rate.
# ===============================================================
missing_rate = 0.0



# ===============================================================
# Target feature type.
# ===============================================================
if target_vae_type == 'Original':
    target_feature_type = 'VarianceTop1' # varianceが0の遺伝子を除く
    # target_feature_type = 'Original' # 全ての遺伝子を使用
elif target_vae_type == 'VAE':
    target_feature_type = 'Epo2000_Lr0.0001_Hid512_256_128_Lat64_Bat64_Dro0.1_ActTanh'
elif target_vae_type == 'VQ_VAE':
    if cell_name == 'All':
    #     target_feature_type = 'Epo500_Lr0.002_Hid1000_512_256_Lat128_Bat64_Comc0.25'
        target_feature_type = 'Epo500_Lr0.0002_Hid512_256_128_Lat64_Bat64_Comc0.25_Lr0.0002_ActSELU_ScaleMaxAbs'
    elif cell_name == 'AllCellParallel' and missing_rate == 0.1:
        target_feature_type = 'Epo500_Lr0.0002_Hid512_1024_512_Lat256_Bat64_Comc0.25_Dro0.1_ActLeakyReLU0.5_Missing0.1_ScaleMaxAbs'
    elif cell_name == 'AllCellParallel' and missing_rate == 0.0:
        target_feature_type = 'Epo500_Lr0.0002_Hid512_1024_512_Lat256_Bat64_Comc0.25_Dro0.1_ActLeakyReLU0.5_Missing0.0_ScaleMaxAbs'
    elif cell_name == 'AllCell' and missing_rate == 0.1:
        target_feature_type = 'Epo500_Lr0.0002_Hid512_256_128_Lat64_Bat64_Comc0.25_Dro0.1_ActLeakyReLU0.5_Missing0.1_ScaleMaxAbs'
    elif cell_name == 'AllCell' and missing_rate == 0.0:
        target_feature_type = 'Epo500_Lr0.0002_Hid512_256_128_Lat64_Bat64_Comc0.25_Dro0.1_ActLeakyReLU0.5_Missing0.0_ScaleMaxAbs'
        
elif target_vae_type == 'SQ_VAE':
    target_feature_type = 'Epo500_Lr0.002_Hid1000_512_256_Lat128_Bat64_Varq5_Temp1e-05_ScaleOrig'
elif target_vae_type == 'GaussianSQVAE_gaussian_1':
    if cell_name == 'All':
        target_feature_type = 'Epo500_Hid512_256_128_Lat64_Bat64_Varq30.0_ScaleMaxAbs'
    elif cell_name == 'AllCellParallel':
        target_feature_type = 'Epo500_Hid512_1024_512_Lat256_Bat64_Varq30.0_ActSELU_Missing0.1_ScaleMaxAbs'
#     elif cell_name == 'AllCellParallel' and pert_type == 'trt_sh.cgs':
#         target_feature_type = 'Epo500_Hid512_1024_512_Lat256_Bat64_Varq50.0_ActTanh_ScaleMaxAbs'
#     elif cell_name == 'AllCellParallel' and pert_type == 'trt_oe':
#         target_feature_type = 'Epo100_Hid512_1024_512_Lat256_Bat64_Varq30.0_ActSELU_ScaleMaxAbs'

    
# ===============================================================
# Disease feature type.
# ===============================================================
# disease_profile_type = 'disease_signature.tabs'
disease_profile_type = 'disease_and_patient_signature.tabs'

if disease_vae_type == 'Original':
    disease_feature_type = 'VarianceTop1' # varianceが0の遺伝子を除く
    # disease_feature_type = 'Original' # 全ての遺伝子を使用
    
elif disease_vae_type == 'VAE':
    if disease_profile_type == 'disease_signature.tabs':
        disease_feature_type = 'Epo2000_Lr0.0001_Hid1024_512_256_Lat128_Bat64_Dro0.2_ScaleMaxAbs_ActReLU'
    elif disease_profile_type == 'disease_and_patient_signature.tabs':
        disease_feature_type = 'Epo2000_Lr0.0001_Hid1024_512_256_Lat128_Bat64_Dro0.2_ScaleMaxAbs_ActReLU'
        
elif disease_vae_type == 'VQ_VAE':
    disease_feature_type = 'Epo2000_Lr0.002_Hid1000_512_256_Lat128_Bat64_Comc0.25_ScaleStd'
    
elif disease_vae_type == 'SQ_VAE':
    disease_feature_type = 'Epo2000_Lr0.002_Hid1000_512_256_Lat128_Bat64_Varq5_Temp1e-05_ScaleMaxAbs'
    
elif disease_vae_type == 'GaussianSQVAE_gaussian_1':
    if disease_profile_type == 'disease_signature.tabs':
        disease_feature_type = 'Epo1000_Hid1000_512_256_Lat128_Bat64_Varq30_Dro0.1_ActTanh_ScaleMaxAbs'
    elif disease_profile_type == 'disease_and_patient_signature.tabs':
        disease_feature_type = 'Epo1000_Hid1000_512_256_Lat128_Bat64_Varq40_Dro0.1_ActTanh_ScaleMaxAbs'
```

```python
import sys  
import torch
import random
import argparse
import numpy as np
import time
import torch.nn as nn
import pandas as pd
import os

# # Import the required files
# sys.path.insert(1, '../scr_train/')
# import utils, ConcatNN, train_gene_nn
# from utils import *
# from ConcatNN import *
# from train_gene_nn import *

# Apply the seed to reproduct the results
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)


# =================================================================================
# Default settings
parser = argparse.ArgumentParser()

# =================================================================================
# Protein signatures.
# =================================================================================

# Pert type.
parser.add_argument('--pert_type', type=str, default=pert_type, 
                    help='Perturbation type of protein signatures, e.g., trt_sh.cgs or trt_oe') # knockdown signatures

# Cell name.
parser.add_argument('--cell_name', type=str, 
                    default=cell_name, 
                    help='Cell name of LINCS files, e.g., mcf7') # cell line of protein signatures

# Target selection.
parser.add_argument('--target_select', type=str, default=target_select, 
                    help='Targets used for training. If supervised learning, only known proteins are used for training.')

# Target feature type.
parser.add_argument('--target_feature_type', type=str, default= target_feature_type, 
                    help='target feature types, e.g., Epo2000_Lr0.0001_Hid512_256_128_Lat64_Bat64_Dro0.1_ActTanh, VarianceTop1 or Original')

# Target expression profile path.
parser.add_argument('--target_expression_file', type=str, default='../../../LINCS/latest_version/imputation/imputed_data/', 
                    help='Path of the training gene expression profile dataset for the NN')

# Target VAE type.
parser.add_argument('--target_vae_type', type=str, default=target_vae_type, help='Protein VAE type, e.g., VAE, VQ_VAE or Original')

# Select Cell lines based on missing rates.
# --cell_name='AllCell'の時のみ使用
parser.add_argument(
    '--cell_missing_rate', type=float, 
    default=0.1,
    help='Threshold of missing rates of cell lines when selecting AllCellParallel as --cell_name'
)


# =================================================================================
# Disease signatures.
# =================================================================================

# Disease profile type.
parser.add_argument('--disease_profile_type', type=str, default=disease_profile_type, 
                    help='Disease signature type, e.g., disease_signature.tabs or patient_signature.tabs') # disease signatures

# Disease selection.
parser.add_argument('--disease_select', type=str, default='79', 
                    help='Diseases used for training. If supervised learning, only known diseases are used for training.')

# Disease feature type.
parser.add_argument('--disease_feature_type', type=str, default=disease_feature_type, 
                    help='disease feature types, e.g., Epo2000_Lr0.0001_Hid512_256_128_Lat64_Bat64_Dro0.2_ScaleStd, VarianceTop1 or Original')

# Disease expression file.
parser.add_argument('--disease_expression_file', type=str, default='../../../CREEDS/processed_data/expression-based_', 
                    help='Path of the training gene expression profile dataset for the VAE')

# Disease VAE type.
parser.add_argument('--disease_vae_type', type=str, default=disease_vae_type, help='Disease VAE type, e.g., VAE, VQ_VAE or Original')

# =================================================================================
# Pairwise type
# =================================================================================
parser.add_argument('--pairwise_type', type=str, default='Concat', help='Feature types of protein–disease pairs, e.g., Concat or DisCosine_TarCosine_Kron')

# =================================================================================
# Hyper-parameters of GeneVAE
# =================================================================================
parser.add_argument("-f", required=False)
args = parser.parse_args()
```

```python
# ====== Make output directory ====== #

def MakeOutputDir(args):
    
    if args.disease_vae_type == 'Original' and args.target_vae_type == 'Original':
        disease_feature_path = f"DiseaseFeaeture_{args.disease_feature_type}"
        target_feature_path = f"TargetFeaeture_{args.target_feature_type}"
    else:
        disease_feature_path = f"DiseaseFeaeture_{args.disease_vae_type}_{args.disease_feature_type}"
        target_feature_path = f"TargetFeaeture_{args.target_vae_type}_{args.target_feature_type}"

    o_base_dir = f"../data/{args.pert_type}/{disease_feature_path}/" # Disease feature type
    o_base_dir = f"{o_base_dir}/{target_feature_path}/" # Target feature type
    o_base_dir = f"{o_base_dir}/CellLine_{args.cell_name}/PairD{args.disease_select}T{args.target_select}" # Cell line
    
    return o_base_dir
```

```python
def show_gene_vae_hyperparamaters(args):

    # Hyper-parameters
    params = {}
    print('\n\nConcatNN Hyperparameter Information:')
    print('='*50)
    params['PERTURBATION_TYPE'] = args.pert_type 
    params['CELL_NAME'] = args.cell_name
    params['TARGET_FEATURE_TYPE'] = args.target_feature_type
    
    params['DISEASE_PROFILE_TYPE'] = args.disease_profile_type
    params['DISEASE_FEATURE_TYPE'] = args.disease_feature_type
    
    params['DISEASE_VAE_TYPE'] = args.disease_vae_type
    params['PROTEIN_VAE_TYPE'] = args.target_vae_type
    
    params['DISEASE_UNLABELED_DATA'] = args.disease_select
    params['TARGET_UNLABELED_DATA'] = args.target_select

    for param in params:
        string = param + ' ' * (5 - len(param))
        print('{}:   {}'.format(string, params[param]))
    print('='*50)
    
    
# ========================================================= #
# Print GeneVAE hyperparameter information
show_gene_vae_hyperparamaters(args)
```

```python
# ===== Label data (Yamato-san) ====== #

# Load gold standard data.
i_f = f"../../semisupervised/data/{args.pert_type}/LabelData/label.txt"
gold_df = pd.read_csv( i_f, sep="\t", encoding = 'shift-jis' )

# Gene and disease list.
gold_genes = sorted(list(set(gold_df['gene']))) # ソートしておく
l_dis_list1 = sorted( list( set(gold_df['disease_id']) ) ) # disease list
l_dis_list2 = [ s + '_01' for s in l_dis_list1] # Patient list

print("標的数：", len(gold_genes))
print( "疾患数：{}".format( len(l_dis_list1) ) )

gold_genes[0:5]
```

```python
# ===== Label data based (Otani-san data) ====== #

# Load gold standard data.
i_f = f"../../semisupervised/data_old//{pert_type}/LabelData/label.txt"
ot_gold_df = pd.read_csv( i_f, sep="\t", encoding = 'shift-jis' )

# Gene and disease list.
ot_gold_genes = sorted(list(set(ot_gold_df['gene']))) # ソートしておく
ot_dis_list = sorted( list( set(ot_gold_df['disease_id']) ) ) # disease list

print('='*50)
print('Otani-san data')
print("標的数：", len(ot_gold_genes))
print( "疾患数：{}".format( len(ot_dis_list) ) )
print('='*50)

ot_gold_genes[0:5]
```

```python
# ====== Target variance data ===== #
if 'Variance' in args.target_feature_type:
    tar_feature_sel_ratio = args.target_feature_type.replace('VarianceTop','') # 1なら分散0以外のものは全部採用, それ以外なら分散の大きい上位x%を採用

    i_f = f"../../semisupervised/data/{args.pert_type}/FeatureSelection/Variance/target_variance_Top{ tar_feature_sel_ratio }.txt"
    tar_var_df = pd.read_csv( i_f, sep = '\t' )

    tar_fea_list = list(set(tar_var_df['gene'])) # Target featrue list

    print(tar_var_df)
```

```python
# ============================================
# Protein data
# ============================================

# ===== Original features ===== #
if args.target_vae_type == 'Original':
    # Original features.
    protein_data = pd.read_csv(args.target_expression_file + args.pert_type +'.txt', sep='\t' )

    # Mean target signatures across cell lines.
    if args.cell_name == "All":
        protein_data = protein_data.drop('cell_mfc_name', 
                              axis = 1
                             ).groupby(by = 'cmap_name'
                             ).mean().reset_index() # average across all cell lines
    else:
        protein_data = protein_data[ protein_data['cell_mfc_name'] == args.cell_name 
                        ].drop('cell_mfc_name', axis = 1) # cellを選択
    protein_data = protein_data.set_index( 'cmap_name' ) # indexを指定
    
    
# ===== VAE features. ===== #
else:
    # VAE features.
    i_f = f'../../VAE/data/{args.target_vae_type}/{args.pert_type}/CellLine_{args.cell_name}/{args.target_feature_type}/latent_vectors.txt'
    if cell_name != 'AllCellParallel':
        protein_data = pd.read_csv( i_f, sep = '\t', index_col='cmap_name')
    else:
        protein_data = pd.read_csv( i_f, sep = '\t', index_col=0)
        
        
# ===== Labeled data ===== #    
# protein_data = protein_data.loc[ gold_genes ] # known proteinsを選択、全ての細胞で並びを揃えるため
l_protein_data = protein_data.loc[ l_pro_list ] # known proteinsを選択、全ての細胞で並びを揃えるため


# ===== Uabeled data ===== #
u_protein_data = protein_data.loc[ u_pro_list ] # Select unlabeled diseas signatures


# ====== 遺伝子を選択 ===== #
if 'Variance' in args.target_feature_type:
    l_protein_data = l_protein_data[tar_fea_list] # geneを選択
    u_protein_data = u_protein_data[tar_fea_list] # geneを選択
    

# ===== 標準化 ===== #
l_protein_data = (l_protein_data - l_protein_data.min()) / (l_protein_data.max() - l_protein_data.min()) # 各特徴量について標準化
l_protein_data = l_protein_data.dropna(how = 'any', axis = 1)
u_protein_data = (u_protein_data - u_protein_data.min()) / (u_protein_data.max() - u_protein_data.min()) # 各特徴量について標準化
u_protein_data = u_protein_data.dropna(how = 'any', axis = 1)


# ===== Feature data.===== #
l_protein_feature_df = pd.DataFrame(list(l_protein_data.columns), columns=['features']) # features used for training.
u_protein_feature_df = pd.DataFrame(list(u_protein_data.columns), columns=['features']) # features used for training.
o_dir = f"{MakeOutputDir(args)}/feature_files/" # Cell line
os.makedirs( o_dir, exist_ok=True ) # Make dirs.
o_f = f"{o_dir}/l_protein_features.txt"
l_protein_feature_df.to_csv( o_f, sep = '\t', index = None)
o_f = f"{o_dir}/u_protein_features.txt"
u_protein_feature_df.to_csv( o_f, sep = '\t', index = None)

print('Labeled disease data: ', l_protein_data.shape) # (nums, dimension)
print('Unlabeled disease data : ', u_protein_data.shape) # (nums, dimension)

l_protein_data
```

```python
# ====== Disease variance data ===== #
if 'Variance' in args.disease_feature_type:
    dis_feature_sel_ratio = args.disease_feature_type.replace('VarianceTop','') # 1なら分散0以外のものは全部採用, それ以外なら分散の大きい上位x%を採用

    i_f = f"../../semisupervised/data/{args.pert_type}/FeatureSelection/Variance/disease_variance_Top{ dis_feature_sel_ratio }.txt"
    dis_var_df = pd.read_csv( i_f, sep = '\t' )

    dis_fea_list = list( set(dis_var_df['gene']) ) # Disease feature list

    print(dis_var_df)
```

```python
# ==================================
# Disease data
# ==================================

# ===== 読み込む ===== #
if args.disease_vae_type == 'Original':
    # Original features.
    all_disease_data = pd.read_csv(args.disease_expression_file + args.disease_profile_type, 
                             sep = '\t',
                             index_col='disease'
                            )
else:
    # VAE features.
    i_f = f'../../VAE/data/{args.disease_vae_type}/disease/{args.disease_profile_type}/{args.disease_feature_type}/latent_vectors.txt'
    all_disease_data = pd.read_csv(i_f, sep = '\t', index_col='disease')
    

    
# ===== Labeled data ===== #    
l_disease_data1 = all_disease_data.loc[ l_dis_list1 ] # Select known diseases.
l_disease_data2 = all_disease_data.loc[ l_dis_list2 ] # Select known disease patients.


# ===== Uabeled data: disease-specific, patient-specific ===== #
u_dis_list = list(set(all_disease_data.index) - set(l_dis_list1)) # Unlabeled disease list
u_dis_list1 = [ s for s in u_dis_list if '_' not in s] # Select disease-specific list
u_dis_list2 = [s+'_01' for s in u_dis_list1] # Select patient 01 id list

u_disease_data1 = all_disease_data.loc[u_dis_list1] # Select unlabeled diseas signatures
u_disease_data2 = all_disease_data.loc[u_dis_list2] # Select unlabeled diseas signatures


# ====== 遺伝子を選択 ===== #
if 'Variance' in args.disease_feature_type:
    l_disease_data1 = l_disease_data1[dis_fea_list] # Select genes.
    l_disease_data2 = l_disease_data2[dis_fea_list] # Select genes.
    u_disease_data1 = u_disease_data1[dis_fea_list] # Select genes.
    u_disease_data2 = u_disease_data2[dis_fea_list] # Select genes.

    
# ===== 標準化 ===== #
l_disease_data1 = (l_disease_data1 - l_disease_data1.min()) / (l_disease_data1.max() - l_disease_data1.min()) # 各特徴量について標準化
l_disease_data1 = l_disease_data1.dropna(how = 'any', axis = 1)
l_disease_data2 = (l_disease_data2 - l_disease_data2.min()) / (l_disease_data2.max() - l_disease_data2.min()) # 各特徴量について標準化
l_disease_data2 = l_disease_data2.dropna(how = 'any', axis = 1)
u_disease_data1 = (u_disease_data1 - u_disease_data1.min()) / (u_disease_data1.max() - u_disease_data1.min()) # 各特徴量について標準化
u_disease_data1 = u_disease_data1.dropna(how = 'any', axis = 1)
u_disease_data2 = (u_disease_data2 - u_disease_data2.min()) / (u_disease_data2.max() - u_disease_data2.min()) # 各特徴量について標準化
u_disease_data2 = u_disease_data2.dropna(how = 'any', axis = 1)


# ===== Feature data. ===== #
l_disease_feature_df1 = pd.DataFrame(list(l_disease_data1.columns), columns=['features'])
l_disease_feature_df2 = pd.DataFrame(list(l_disease_data2.columns), columns=['features'])
u_disease_feature_df1 = pd.DataFrame(list(u_disease_data1.columns), columns=['features'])
u_disease_feature_df2 = pd.DataFrame(list(u_disease_data2.columns), columns=['features'])
o_dir = f"{MakeOutputDir(args)}/feature_files/" # Cell line
os.makedirs( o_dir, exist_ok=True ) # Make dirs.
o_f = f"{o_dir}/l_disease_features1.txt"
l_disease_feature_df1.to_csv( o_f, sep = '\t', index = False )
o_f = f"{o_dir}/l_disease_features2.txt"
l_disease_feature_df2.to_csv( o_f, sep = '\t', index = False )
o_f = f"{o_dir}/u_disease_features1.txt"
u_disease_feature_df1.to_csv( o_f, sep = '\t', index = False )
o_f = f"{o_dir}/u_disease_features2.txt"
u_disease_feature_df2.to_csv( o_f, sep = '\t', index = False )

print('Labeled disease data: ', l_disease_data1.shape) # (nums, dimension)
print('Labeled patient data: ', l_disease_data2.shape) # (nums, dimension)
print('Unlabeled disease data1 : ', u_disease_data1.shape) # (nums, dimension)
print('Unlabeled patient data2 : ', u_disease_data2.shape) # (nums, dimension)
l_disease_data1
```

```python
# =================================================================================
# 疾患とタンパクのデータをマージする
# =================================================================================
# Concat protein signatures and disease signatures.

# ------ Target signaturesの繰り返し ------ #
l_protein_tensor = l_protein_data.values # ラベルありタンパク => ラベルありデータ用
l_protein_tensor = np.tile(l_protein_tensor, (len(l_dis_list1), 1)) # protein 1, protein 2, ...., protein 1, protein 2, ....
u_protein_known_tensor = l_protein_data.values # ラベルありタンパク => ラベルなしデータ用
u_protein_known_tensor = np.tile(u_protein_known_tensor, (len(u_dis_list1), 1)) # protein 1, protein 2, ...., protein 1, protein 2, ....
u_protein_orphan_tensor = u_protein_data.values # ラベルなしタンパク => ラベルなしデータ用
u_protein_orphan_tensor = np.tile(u_protein_orphan_tensor, (len(l_dis_list1), 1)) # protein 1, protein 2, ...., protein 1, protein 2, ....

# ----- Disease signatureの繰り返し ------ #
l_disease_tensor = l_disease_data1.values # ラベルあり疾患 => ラベルありデータ用
l_disease_tensor = np.repeat( l_disease_tensor, len( l_pro_list ), axis=0 ) # disease 1, disease 1, ...., disease 2, disease 2, ....
u_disease_known_tensor1 = l_disease_data1.values # ラベルあり疾患 => ラベルなしデータ用
u_disease_known_tensor1 = np.repeat( u_disease_known_tensor1, len( u_pro_list ), axis=0 ) # disease 1, disease 1, ...., disease 2, disease 2, ....
u_disease_orphan_tensor1 = u_disease_data1.values # ラベルなし疾患 => ラベルなしデータ用
u_disease_orphan_tensor1 = np.repeat( u_disease_orphan_tensor1, len( l_pro_list ), axis=0 ) # disease 1, disease 1, ...., disease 2, disease 2, ....

# ----- Patient signatureの繰り返し ----- #
u_disease_known_tensor2 = l_disease_data2.values # ラベルあり疾患 => ラベルなしデータ用
u_disease_known_tensor2 = np.repeat( u_disease_known_tensor2, len( u_pro_list ), axis=0 ) # disease 1, disease 1, ...., disease 2, disease 2, ....
u_disease_orphan_tensor2 = u_disease_data2.values # ラベルなし疾患 => ラベルなしデータ用
u_disease_orphan_tensor2 = np.repeat( u_disease_orphan_tensor2, len( l_pro_list ), axis=0 ) # disease 1, disease 1, ...., disease 2, disease 2, ....


# ----- Known diseaseとKnown proteinsのペアのKernel行列のindexを作成 ------ #
l_dis_pair_list = list( np.repeat( l_dis_list1, len( l_pro_list ) ) ) # disease 1, disease 1, ...., disease 2, disease 2, ....
l_tar_pair_list = l_pro_list * len( l_dis_list1 ) # protein 1, protein 2, ...., protein 1, protein 2, ....
l_pair_index_list = [ str(d) + '__' + str(t) for d,t in zip( l_dis_pair_list, l_tar_pair_list ) ] # diseaseとtargetのpairリスト

# ----- Orphan diseaseとKnown proteinのペアのKernel行列のindexを作成 ------ #
u_dis_orphan_pair_list = list( np.repeat( u_dis_list1, len( l_pro_list ) ) ) # disease 1, disease 1, ...., disease 2, disease 2, ....
u_tar_known_pair_list = l_pro_list * len( u_dis_list1 ) # protein 1, protein 2, ...., protein 1, protein 2, ....
u_pair_dis_orphan_index_list = [ str(d) + '__' + str(t) for d,t in zip( u_dis_orphan_pair_list, u_tar_known_pair_list ) ] # diseaseとtargetのpairリスト

# ----- Known diseaseとOrphan proteinsのペアのKernel行列のindexを作成 ------ #
u_dis_known_pair_list = list( np.repeat( l_dis_list1, len( u_pro_list ) ) ) # disease 1, disease 1, ...., disease 2, disease 2, ....
u_tar_orphan_pair_list = u_pro_list * len( l_dis_list1 ) # protein 1, protein 2, ...., protein 1, protein 2, ....
u_pair_pro_orphan_index_list = [ str(d) + '__' + str(t) for d,t in zip( u_dis_known_pair_list, u_tar_orphan_pair_list ) ] # diseaseとtargetのpairリスト


# ===== Merge disease and protein data ===== #
# ラベルありデータ：　ラベルあり疾患 x ラベルありタンパク
l_mg_tensor = np.concatenate( [l_disease_tensor, l_protein_tensor  ], 1) # merge
l_data = pd.DataFrame(l_mg_tensor, 
             columns= list(l_disease_data1.columns) + list(l_protein_data.columns),
             index = l_pair_index_list
            )

# ラベルなしデータ：　 ラベルなし疾患 x ラベルありタンパク
u_mg_dis_orphan_tensor1 = np.concatenate( [u_disease_orphan_tensor1, u_protein_known_tensor], 1) # merge
u_dis_orphan_data1 = pd.DataFrame(u_mg_dis_orphan_tensor1, 
             columns= list(u_disease_data1.columns) + list(l_protein_data.columns),
             index = u_pair_dis_orphan_index_list
            )

# ラベルなしデータ：　ラベルあり疾患 x ラベルなしタンパク
u_mg_pro_orphan_tensor1 = np.concatenate( [u_disease_known_tensor1, u_protein_orphan_tensor], 1) # merge
u_pro_orphan_data1 = pd.DataFrame(u_mg_pro_orphan_tensor1, 
             columns= list(l_disease_data1.columns) + list(u_protein_data.columns),
             index = u_pair_pro_orphan_index_list
            )

# ラベルなしデータ：　 ラベルなし患者 x ラベルありタンパク
u_mg_dis_orphan_tensor2 = np.concatenate( [u_disease_orphan_tensor2, u_protein_known_tensor], 1) # merge
u_dis_orphan_data2 = pd.DataFrame(u_mg_dis_orphan_tensor2, 
             columns= list(u_disease_data2.columns) + list(l_protein_data.columns),
             index = u_pair_dis_orphan_index_list
            )

# ラベルなしデータ：　ラベルあり患者 x ラベルなしタンパク
u_mg_pro_orphan_tensor2 = np.concatenate( [u_disease_known_tensor2, u_protein_orphan_tensor], 1) # merge
u_pro_orphan_data2 = pd.DataFrame(u_mg_pro_orphan_tensor2, 
             columns= list(l_disease_data2.columns) + list(u_protein_data.columns),
             index = u_pair_pro_orphan_index_list
            )

print("ラベルありの疾患 x ラベルなしのタンパク：{}".format(len(l_dis_list1) * len(u_pro_list)))
print("ラベルありの疾患 x ラベルなしのタンパク：{}".format(len(u_pro_orphan_data1)))
print("ラベルありの疾患 x ラベルなしのタンパク：{}".format(len(u_pro_orphan_data2)))

print("ラベルなしの疾患 x ラベルありのタンパク：{}".format(len(u_dis_list1) * len(l_pro_list)))
print("ラベルなしの疾患 x ラベルありのタンパク：{}".format(len(u_dis_orphan_data1)))
print("ラベルなしの疾患 x ラベルありのタンパク：{}".format(len(u_dis_orphan_data2)))


# ----- (ラベルありの疾患 x ラベルなしのタンパク)と(ラベルなしの疾患 x ラベルありのタンパク)をマージ ----- #
u_data1 = pd.concat([u_pro_orphan_data1, u_dis_orphan_data1], axis=0) # 疾患
u_data2 = pd.concat([u_pro_orphan_data2, u_dis_orphan_data2], axis=0) # 患者


# ----- Save ----- #
if args.target_vae_type == 'Original' and args.disease_vae_type == 'Original':
    disease_feature_path = f"DiseaseFeaeture_{args.disease_feature_type}"
    target_feature_path = f"TargetFeaeture_{args.target_feature_type}"
else:
    disease_feature_path = f"DiseaseFeaeture_{args.disease_vae_type}_{args.disease_feature_type}"
    target_feature_path = f"TargetFeaeture_{args.target_vae_type}_{args.target_feature_type}"
        
o_base_dir = f"../data/{args.pert_type}/{disease_feature_path}/" # Disease feature type
o_base_dir = f"{o_base_dir}/{target_feature_path}/" # Target feature type
# o_base_dir = f"../data/{args.pert_type}/DiseaseFeaeture_{args.vae_type}_{args.disease_feature_type}/" # Disease feature type
# o_base_dir = f"{o_base_dir}/TargetFeaeture_{args.vae_type}_{args.target_feature_type}/" # Target feature type
    
o_dir = f"{o_base_dir}/CellLine_{args.cell_name}/" # Cell line
o_dir = f"{o_dir}/PairD{args.disease_select}T{args.target_select}/feature_files/" # Unlabeled data
os.makedirs( o_dir, exist_ok=True ) # Make dirs.
o_f = f"{o_dir}/l_feature_expression_cv.tabs"
l_data.to_csv( o_f, sep = '\t', index = True)
o_f = f"{o_dir}/u1_feature_expression_cv.tabs"
u_data1.to_csv( o_f, sep = '\t', index = True)
o_f = f"{o_dir}/u2_feature_expression_cv.tabs"
u_data2.to_csv( o_f, sep = '\t', index = True)


print( "Size of labeled protein–disease signatures: {}".format(l_data.shape) )
print( "Size of unlabeled protein–disease signatures 1: {}".format(u_data1.shape) )
print( "Size of unlabeled protein–disease signatures 2: {}".format(u_data2.shape) )
l_data
```

```python
# ===== クロスバリデーション用のデータを作成 ====== #

# train_x: ペアワイズクロスバリデーションを行うので、どのfoldでもdataを使用。
# train_y: matrix型からdataframe型に変換。
# test_x: data (train_x) からtest_yに含まれるペアのみを抽出。
# test_y: matrixがたからdataframe型に変換。

for i in range(1, 6):

    # ===== Training label data ====== #
    i_train_y = f"../../semisupervised/data/{args.pert_type}/performance_evaluation/CellLine_All/training_files/training_label_foldf{i}.tabs"
    train_y = pd.read_csv( i_train_y, 
                          sep = '\t' 
                         ).set_index("Unnamed: 0") # set index.
    train_y = train_y.stack().reset_index() # 横持ち->縦持ち
    train_y.columns = ['gene', 'disease_id', 'label' ] # Column name.
    train_y['pair_index'] = [ d + '__' + g 
                             for d,g in zip(train_y['disease_id'], train_y['gene']) ] # ペアの列を作成
    train_y = train_y.set_index('pair_index') # set index.
    train_y = train_y.loc[l_pair_index_list] # feature dataと揃える
    train_y = train_y.drop( ['gene', 'disease_id'], axis = 1 ) # remove columns.

    # ----- Save ----- #
    o_dir = f"{o_base_dir}/CellLine_{args.cell_name}/PairD{args.disease_select}T{args.target_select}/training_files/" # Cell line
    os.makedirs( o_dir, exist_ok=True ) # Make dirs.
    o_f = f"{o_dir}/training_label_foldf{i}.tabs"
    train_y.to_csv( o_f, sep = '\t', index = True)
    
    
    # ===== Test label data ===== #
    i_test_y = f"../../semisupervised/data/{args.pert_type}/performance_evaluation/CellLine_All/test_files/test_label_foldf{i}.tabs"
    test_y = pd.read_csv( i_test_y, 
                          sep = '\t' 
                         ).set_index( "Unnamed: 0" ) # indexを指定
    test_y = test_y.stack().reset_index() # 横持ち->縦持ち
    test_y.columns = ['gene', 'disease_id', 'label' ] # 列名変更
    test_y['pair_index'] = [ d + '__' + g 
                       for d,g in zip(test_y['disease_id'], test_y['gene']) ] # ペアの列を作成
    test_y = test_y.set_index('pair_index') # set index.
    test_y = test_y.drop( ['gene', 'disease_id'], axis=1 ) # remove columns.

    # ----- Save ----- #
    o_dir = f"{o_base_dir}/CellLine_{args.cell_name}/PairD{args.disease_select}T{args.target_select}/test_files/" # Cell line
    os.makedirs( o_dir, exist_ok=True ) # Make dirs.
    o_f = f"{o_dir}/test_label_foldf{i}.tabs"
    test_y.to_csv( o_f, sep = '\t', index = True)


    # ====== Test feature data ====== #
    test_y_index = list(test_y.index) # index of test label data
    test_x = l_data.loc[test_y_index] # extract test pairs

    # ----- Save ----- #
    o_dir = f"{o_base_dir}/CellLine_{args.cell_name}/PairD{args.disease_select}T{args.target_select}/test_files/" # Cell line
    os.makedirs( o_dir, exist_ok=True ) # Make dirs.
    o_f = f"{o_dir}/test_feature_expression_foldf{i}.tabs"
    test_x.to_csv( o_f, sep = '\t', index = True)

train_y
```
